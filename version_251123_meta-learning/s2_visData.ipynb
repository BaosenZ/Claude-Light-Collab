```python
import pandas as pd

df = pd.read_csv('data/data_full.csv')
print(df.shape)
df.head(2)
```

```python
df['date'].value_counts()
```

```python
# Draw plots at 0.5, 0.5, 0.5 at date to see the changes
df_all0_5 = df[(df['R']==0.5) & (df['G']==0.5) & (df['B']==0.5) & (df['time_of_day']<12)].copy()
df_all0_5['date_str'] = pd.to_datetime(df_all0_5['date'].astype(str), format='%Y%m%d')
df_all0_5
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(5, 3))
plt.plot(df_all0_5['date_str'], df_all0_5['I_445'], marker='o', linestyle='-', label='I_445')
plt.plot(df_all0_5['date_str'], df_all0_5['I_515'], marker='o', linestyle='-', label='I_515')
plt.plot(df_all0_5['date_str'], df_all0_5['I_630'], marker='o', linestyle='-', label='I_630')
plt.plot(df_all0_5['date_str'], df_all0_5['I_clear'], marker='o', linestyle='-', label='I_clear')

plt.xticks(rotation=45, ha='right')
plt.xlabel('Date')
plt.ylabel('I (R=0.5,G=0.5,B=0.5)')
plt.legend()

plt.tight_layout()
plt.show()
```
